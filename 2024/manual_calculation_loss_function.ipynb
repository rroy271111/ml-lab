```python
import torch
import torch.nn.functional as F

# True class (index of correct label)
true_class_index = torch.tensor([1]) # dog

# Model's predicted logits (raw scores before softmax)
logits = torch.tensor([[2.0, 3.5, 0.5]]) # dog, cat, bird ( shape : (1,3))


# Compute softmax probabilities
probs = F.softmax(logits, dim=1)
print("Softmax probs:", probs)

# Manual Cross Entropy Loss
log_prob_true = -torch.log(probs[0, true_class_index])
print("Manual Cross Entropy Loss:", log_prob_true)

# Built-in Cross Entropy Loss (automatically applies softmax + log + loss)
loss = F.cross_entropy(logits, true_class_index)
print("Python Cross Entropy Loss:", loss)
```
