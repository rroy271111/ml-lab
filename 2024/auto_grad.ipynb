```python
import torch

x = torch.randn(3, requires_grad=True)
print(x)

y = x + 2
print(y)

z = y * y * 2
z = z.mean()
print("z ",z)
print("############")
v = torch.tensor([0.1,1.0, 0.001], dtype=torch.float32)
print("v ",v)
z.backward() #dz/dx
print("z ",z)
print("x ",x.grad)


############# autograd demo
second = torch.tensor([2.0], requires_grad=True)
first = second**2
first.backward()
print(second.grad)
```
