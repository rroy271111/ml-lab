```python
import torch

# Data (inputs and outputs)
X = torch.tensor([1, 2, 3, 4], dtype=torch.float32)
Y = torch.tensor([2, 4, 6, 8], dtype=torch.float32)

# Initialize weight with requires_grad set to True for gradient tracking
weight = torch.tensor(0.0, dtype=torch.float32, requires_grad=True)

# Model prediction: y = weight * x
def forward(x):
    return weight * x

# Loss function: Mean Squared Error (MSE)
def loss(y, y_predicted):
    return ((y_predicted - y) ** 2).mean()

# Before training prediction
print(f'Prediction before training: f(5) = {forward(5).item():.3f}')

# Training loop
learning_rate = 0.01
n_iters = 100

for epoch in range(n_iters):
    # Forward pass: compute prediction
    y_pred = forward(X)

    # Compute loss
    l = loss(Y, y_pred)

    # Backward pass: compute gradients
    l.backward()  # This computes dl/dw and stores in weight.grad

    # Update weight
    with torch.no_grad():
        weight -= learning_rate * weight.grad  # Use gradient to update weight

    # Zero gradients after update
    weight.grad.zero_()

    # Print progress every 10 epochs
    if epoch % 10 == 0:
        print(f'epoch {epoch + 1}: weight = {weight.item():.3f}, loss = {l.item():.8f}')

# After training prediction
print(f'Prediction after training: f(100) = {forward(100).item():.3f}')
```
