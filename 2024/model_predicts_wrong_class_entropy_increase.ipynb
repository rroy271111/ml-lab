```python
# Model is Confident but wrong, entropy loss is high

import torch
import torch.nn.functional as F

# True label: Dog (index 1)
true_class_index = torch.tensor([1]) # class 1 = Dog

# Model's (bad) logits: predicts Cat with high confidence
logits = torch.tensor([[5.0, 3.5, -1.0]]) # cat, dog, bird ( shape : (1,3))

# Softmax probabilities
probs = F.softmax(logits, dim=1)
print("softmax probs:", probs)  

# Manual Cross Entropy Loss
manual_loss = -torch.log(probs[0, true_class_index])
print("Manual Cross Entropy Loss:", manual_loss.item())

# Built-in Cross Entropy Loss (automatically applies softmax + log + loss)
loss = F.cross_entropy(logits, true_class_index)
print("Python Cross Entropy Loss:", loss.item())
```
