```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create input values
x = np.linspace(-10, 10, 200)   # 200 points from -10 to 10

# Step 2: Apply ReLU function to input values
y = np.maximum(0,x)

# Step 3: Plot 
plt.figure(figsize=(6,4))
plt.plot(x, y, label='ReLU', color="blue", linewidth=2)
plt.axvline(0, color="black", linestyle="--", linewidth=1)
plt.axhline(0, color="black", linewidth=1)
plt.title("ReLU Activation Function")
plt.xlabel("Input (x)")
plt.ylabel("Output f(x)")
plt.legend()
plt.grid(True)
plt.show()
 
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Original list
my_list = [-5, -3, -1, 0, 1, 3, 5]

# Step 2: Convert to NumPy array for vectorized math
x = np.array(my_list)

# Step 3: Apply ReLU: f(x) = max(0, x)
relu_output = np.maximum(0, x)

# Step 4: Plot input vs output
fig, ax = plt.subplots(figsize=(6, 4))

# Original list values (before ReLU)
ax.plot(x, x, 'o--', label="Original (y = x)", color="red")

# ReLU-transformed values
ax.plot(x, relu_output, 'o-', label="After ReLU", color="blue")

# Labels and grid
ax.set_title("List  ReLU Transformation")
ax.set_xlabel("Input values")
ax.set_ylabel("Output values")
ax.axhline(0, color="black", linewidth=1)  # x-axis
ax.axvline(0, color="black", linewidth=1)  # y-axis
ax.grid(True)
ax.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create random input values
np.random.seed(43)  # For reproducibility
x = np.random.uniform(-10,10,20) # 20 random values between -10 and 10

# Step 2: Apply ReLU function to the input values
relu_output = np.maximum(0,x)   

# Step 3: Plot before vs after
fig, ax = plt.subplots(figsize=(6,4))

# Scatter plot: inputs
ax.scatter(x, x , color='black', label="Original ( y = x)", s = 60 )

# Scatter plot: outputs
ax.scatter(x, relu_output, color='blue', label="After ReLU", s=60)
```
